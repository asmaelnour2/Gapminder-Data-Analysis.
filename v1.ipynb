```python
import pandas as pd
import plotly.express as px
```

```python
df = px.data.gapminder()
```

```python
df[df['country'] == 'India']
```

```python
df
```

```python
df2=df[df['year'] == 2007]
```

```python
px.scatter(data_frame=df2,x='gdpPercap',y='lifeExp',size='pop',color='continent',title='GDP Per Capita vs Life Expectancy in 2007',labels={'gdpPercap':'GDP Per Capita','lifeExp':'Life Expectancy'},template='plotly_dark')
```

```python
df_egypt=df[df['country'] == 'Egypt']   
```

```python
px.line(data_frame=df_egypt,x='year',y='lifeExp',title='Life Expectancy in Egypt',markers=True,labels={'lifeExp':'Life Expectancy','year':'Year'},template='plotly_dark')
```

```python
df_tips = px.data.tips()
df_tips.head()
```

```python
px.pie(data_frame=df_tips,names='day',values='total_bill',title='Total Bill Distribution by Day',template='plotly_dark',labels={'day':'Day of the Week','total_bill':'Total Bill'},color_discrete_sequence=px.colors.qualitative.Pastel,hole=0.6)    
```

```python
px.bar(data_frame=df_tips,x='smoker',y='total_bill',color='day',title='Total Bill by Smoker Status and Day',labels={'smoker':'Smoker Status','total_bill':'Total Bill','day':'Day of the Week'},template='plotly_dark')
```

```python
px.bar(data_frame=df_tips,x='day',y='total_bill',color='smoker',title='Total Bill by Day and Smoker Status',labels={'day':'Day of the Week','total_bill':'Total Bill','smoker':'Smoker Status'},template='plotly_dark') 
```

```python
temp= df_tips.groupby('day')['total_bill'].sum().reset_index()
```

```python
temp
```

```python
px.bar(data_frame=temp,x='day',y='total_bill',title='Total Bill by Day and Smoker Status',labels={'day':'Day of the Week','total_bill':'Total Bill','smoker':'Smoker Status'},template='plotly_dark') 
```
